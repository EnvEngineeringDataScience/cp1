```python
import pandas as pd

dframe=pd.read_html(wiki_link, 
            header=0,
            flavor='bs4')
```

```python
type(dframe)
```

```python
#queremos utilizar la tabla número 3
countriesDF = dframe [2]
```

```python
countriesDF
```

```python
len(countriesDF)
```

```python
countries = countriesDF.copy()
```

```python
countries.columns.to_list()
```

```python
#Quisiera eliminar las columnas del ranking de las naciones unidas
borrar_columna1 = ['United Nations[15]']
countries.drop(columns=borrar_columna1, inplace=True)

borrar_columna2 = ['United Nations[15].1']
countries.drop(columns=borrar_columna2, inplace=True)
```

```python
countries
```

```python
#Cambiaré el nombre de las columnas, las traduciré al español, se añadirán unidades
countries.rename(columns={'Country/Territory':'Pais','UN Region':'Continente',
                         'IMF[1][13]':'FMIestimado', 'IMF[1][13].1':'FMIaño', 'World Bank[14]':'BMestimado', 'World Bank[14].1':'BMaño'},inplace=True)
```

```python
#Estamos realizando un análisis de los PAÍSES 
#Quiero borrar la fila 0 y 1, ya que no forman parte de países o territorios específicos
borrarFilas=[0,1]
countries.drop(index=borrarFilas, inplace = True)
```

```python
countries
```

```python
#Reiniciamos index 
countries.reset_index(drop=True,inplace=True)
```

```python
countries
```

```python
countries.info()
```

```python
#Observamos que las celdas de las dos primeras columnas son del tipo Object
#Verificamos que no hay ningún espacio alrededor de los países
countries.Pais=countries.Pais.str.strip()
```

```python
#Ahora procedemos a verificar las columnas donde hayan caracteres más allá de que sean alfanuméricos o espacios
countries.Pais[countries.Pais.str.contains(pat=r'[^\w\s]', regex=True)]
```

```python
#Verificaremos si hay algún número en las celdas
countries.Pais[countries.Pais.str.contains(pat=r'\d',regex=True)]
```

```python
#La primera columna está limpia
#Ahora analizaremos la segunda columna
#Primero quitaremos, en caso haya, espacios alrededor de los continentes
countries.Continente=countries.Continente.str.strip()
```

```python
#Identificaremos, en caso haya, algún caracter que no sea alfanumérico
countries.Continente[countries.Continente.str.contains(pat=r'[^\w]', regex=True)]
```

```python
countries
```

```python
#De igual manera, veremos si las celdas tienen algún número
countries.Continente[countries.Continente.str.contains(pat=r'\d',regex=True)]
```

```python
#Las dos primeras columnas están limpias
#Solo tienen letras, espacios y símbolos pertenecientes al nombre propio de estos
#Ahora analizaremos las siguientes -> estas tienen números
countries
```

```python
#Nos desharemos de los espacios alrededor
countries.FMIestimado=countries.FMIestimado.str.strip()
```

```python
#¿Las celdas de las siguientes columnas tendrán caracteres aparte de números?
countries.FMIestimado[countries.FMIestimado.str.contains(pat=r'\D',regex=True)]
```

```python
countries.FMIestimado=countries.FMIestimado.str.extract(pat=r'(\d+)',expand=True)
```

```python
countries
```

```python
#La tercera columna está limpia
#Repetimos el proceso con las demás
countries.FMIaño=countries.FMIaño.str.strip()
```

```python
#Ahora veamos en FMIaño si hay algo que no sea un número
countries.FMIaño[countries.FMIaño.str.contains(pat=r'\D',regex=True)]
```

```python
#Para quitar lo que está dentro de los corchetes primero debemos deshacernos de los números que están dentro de estos
#Con el fin de luego extraer solo los números que deseamos
#Entonces, los números dentro del corchete, los cuales están detrás de un espacio, son reemplazados por ''
countries.FMIaño=countries.FMIaño.str.replace(pat=r'(\s\d+)',repl='',regex=True)
```

```python
countries.FMIaño
```

```python
#Como ya nos deshicimos de ese número, ahora extraemos los que nos interesan
countries.FMIaño=countries.FMIaño.str.extract(pat=r'(\d+)', expand=True)
```

```python
countries
```

```python
#Prosigamos con la siguiente columna
countries.BMestimado[countries.BMestimado.str.contains(pat=r'\D',regex=True)]
```

```python
#Vaciaremos estas celdas
countries.BMestimado=countries.BMestimado.str.extract(pat=r'(\d+)',expand=True)
```

```python
countries
```

```python
#Ahora limpiaremos la última línea
#Aseguramos que no hayan espacios alrededor de los números
countries.BMaño=countries.BMaño.str.strip()
```

```python
#¿Qué celdas no tienen solo números?
countries.BMaño[countries.BMaño.str.contains(pat=r'\D',regex=True)]
```

```python
#Nos deshacemos de los números dentro del corchete
countries.BMaño=countries.BMaño.str.replace(pat=r'\s\d+',repl='',regex=True)
```

```python
#Extraemos los números que nos interesan
countries.BMaño=countries.BMaño.str.extract(pat=r'(\d+)',expand=True)
```
